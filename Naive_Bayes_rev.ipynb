#Simple Naive Bayes Code

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer

# Training data
emails = [
    "free win money",
    "free offer",
    "win prize",
    "hello friend",
    "how are you",
    "let us meet"
]

labels = [1, 1, 1, 0, 0, 0]  # 1 = Spam, 0 = Not Spam

# Convert text to numbers
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(emails)

# Train model
model = MultinomialNB()
model.fit(X, labels)

# Test new email
test_email = ["free prize"]
test_vec = vectorizer.transform(test_email)

# Predict
prediction = model.predict(test_vec)

if prediction[0] == 1:
    print("Spam Email")
else:
    print("Not Spam")
```

```python
from sklearn.naive_bayes import BernoulliNB

# Features: [Fever, Cough, Headache]
# 1 = Yes, 0 = No
X = [
    [1, 1, 1],  # Flu
    [1, 1, 0],  # Flu
    [0, 1, 0],  # Cold
    [0, 0, 0],  # Normal
]

# Labels: 0 = Normal, 1 = Cold, 2 = Flu
y = [2, 2, 1, 0]

model = BernoulliNB()
model.fit(X, y)

# New patient: Fever=1, Cough=1, Headache=0
prediction = model.predict([[1, 1, 0]])

print(prediction)
```

#: Weather Prediction

```python
from sklearn.naive_bayes import GaussianNB

# [Temperature, Humidity]
X = [
    [30, 70],  # Play
    [28, 90],  # No
    [25, 65],  # Play
    [20, 95],  # No
]

y = [1,0,1,0]  # 1=Play, 0=No

model = GaussianNB()
model.fit(X, y)

print(model.predict([[27, 68]]))
```

#Sentiment Analysis

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer

reviews = [
    "good product",
    "excellent quality",
    "bad product",
    "worst quality"
]

labels = [1,1,0,0]  # 1=Positive, 0=Negative

vec = CountVectorizer()
X = vec.fit_transform(reviews)

model = MultinomialNB()
model.fit(X, labels)

print(model.predict(vec.transform(["good quality"])))
```


Data Type	Naive Bayes

Yes/No	Bernoulli
